```python
from scipy.stats import ttest_1samp, shapiro, levene, ttest_ind, mannwhitneyu,pearsonr, spearmanr, kendalltau, f_oneway, kruskal


def ab(df_ctrl, df_tst, col):
    # Normal Distribution Assumption Control 
    # HO : It fits Normal Distribution Assumption # H1 : Not.
    test_stat1, pvalue1 = shapiro(df_ctrl[col])
    test_stat2, pvalue2 = shapiro(df_tst[col])

    # Variance Homogenity Assumption Control
    # H0: Homogenous # H1: Not
    test_stat3, pvalue3 = levene(df_ctrl[col], df_tst[col])

    print("df_ctrl", "Normal Distribution Test: ", 'Test Stat = %.4f, p-value = %.4f' % (test_stat1, pvalue1), end=" ---->")
    if pvalue1 < 0.05:
        print("df_ctrl is NOT a Normal Distribution")
    else:
        print("df_ctrl is Normal Distribution")
    print("df_tst", "Normal Distribution Test: ", 'Test Stat = %.4f, p-value = %.4f' % (test_stat2, pvalue2), end=" ---->")
    if pvalue2 < 0.05:
        print("df_tst  is NOT a Normal Distribution")
    else:
        print("df_tst is Normal Distribution.")
    print("df_ctrl", "df_tst", "Variance Homogenity Assumption Control: ", 'Test Stat = %.4f, p-value = %.4f' % (test_stat3, pvalue3), end=" ---->")
    if pvalue3 < 0.05:
        print("df_ctrl-df_tst is NOT Variance Homogenity")
    else:
        print("df_ctrl-df_tst is Variance Homogenity")

    #AB Test here.
    # Parametric (independent 2 sampling t-test)
    if (pvalue1 < 0.05 or pvalue2 < 0.05): #if not a Normal Distribution use mannwhit
        test_stat_final, pvalue_final = mannwhitneyu(df_ctrl[col], df_tst[col])
        print('Test Stat = %.4f, p-value = %.4f' % (test_stat_final, pvalue_final), end=" ---->")
        if pvalue_final < 0.05:
            print('REJECT HO Hypothesis. So we can say there is a statistically significant difference between these two')
        else:
            print('CAN NOT REJECT HO Hypothesis. So we can say there is NOT such a statistically significant difference between these two')

    elif (pvalue1 >= 0.05 and pvalue2 >= 0.05): #if Normal Distribution
        if pvalue3 < 0.05: # NOT HOMOGENOUS. Argumant False
            test_stat_final, pvalue_final = ttest_ind(df_ctrl[col], df_tst[col], equal_var=False)
            print('Test Stat = %.4f, p-value = %.4f' % (test_stat_final, pvalue_final), end=" ---->")
            if pvalue_final < 0.05:
                print('REJECT HO Hypothesis. So we can say there is a statistically significant difference between these two')
            else:
                print('CAN NOT REJECT HO Hypothesis.  So we can say there is NOT such a statistically significant difference between these two')

        else:  # if homogeous pvalue3>=0.05
            test_stat_final, pvalue_final = ttest_ind(df_ctrl[col], df_tst[col], equal_var=True)
            print('Test Stat = %.4f, p-value = %.4f' % (test_stat_final, pvalue_final), end=" ---->")
            if pvalue_final < 0.05:
                print('REJECT HO Hypothesis. So we can say there is a statistically significant difference between these two')
            else:
                print('CAN NOT REJECT HO Hypothesis.  So we can say there is NOT such a statistically significant difference between these two')

    else:  # NON-parametrik all others :S
        test_stat_final, pvalue_final = mannwhitneyu(df_ctrl[col], df_tst[col])
        print('Test Stat = %.4f, p-value = %.4f' % (test_stat_final, pvalue_final), end=" ---->")
        print("ERROR !!!")
        print("pvalue1: ", pvalue1, "pvalue2: ", pvalue2, "pvalue3: ", pvalue3)
```
